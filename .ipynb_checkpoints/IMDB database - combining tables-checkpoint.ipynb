## Importing the different data tables

```python
title_basics_path = "./title.basics.tsv/data.tsv"
title_crew_path = "./title.crew.tsv/data.tsv"
title_principals_path = "./title.principals.tsv/data.tsv"
title_ratings_path = "./title.ratings.tsv/data.tsv"

name_basics_path = "./name.basics.tsv/data.tsv"

basics = pd.read_csv(title_basics_path, sep='\t', index_col="tconst",  na_values=['\\N'], dtype={'endYear':object})
crew = pd.read_csv(title_crew_path, sep='\t', index_col="tconst",  na_values=['\\N'])
principals = pd.read_csv(title_principals_path, sep='\t', index_col="tconst",  na_values=['\\N'])
ratings = pd.read_csv(title_ratings_path, sep='\t', index_col="tconst",  na_values=['\\N'])
names = pd.read_csv(name_basics_path, sep='\t', index_col="nconst",  na_values=['\\N'])
```

```python
print("basics shape: ",basics.shape)
print("crew shape: ",crew.shape)
print("principals shape: ",principals.shape)
print("ratings shape: ",ratings.shape)
print("names shape: ",names.shape)
```

```python
basics.groupby(['titleType']).count()
```

```python
movies_basics = basics[basics.titleType=="movie"]
```

```python
movies_basics.shape
```

```python
movies_basics = movies_basics[movies_basics.isAdult==0]
```

```python
movies_basics.drop(columns=['titleType','isAdult','endYear'], inplace=True)
```

```python
movies_basics_trunc = movies_basics.dropna(axis=0,subset=['startYear'],how='any')
```

```python
movies_basics_trunc.isna().sum()
```

```python
movies_basics_trunc.describe()
```

```python
plot1 = plt.hist(movies_basics_trunc.startYear,bins=50)
```

```python
modern_movies = movies_basics_trunc[movies_basics_trunc.startYear>1930]
modern_movies = modern_movies[modern_movies.startYear<=2020]
```

```python
modern_movies.shape
```

## The Crew Database

```python
crew.shape
```

```python
crew.tail()
```

```python
modern_CM = modern_movies.merge(crew,on="tconst",how="left")
```

```python
modern_CM.shape
```

```python
modern_CM.isna().sum()
```

## The Rating Database

```python
ratings.shape
```

```python
ratings.tail(10)
```

```python
modern_CMR = modern_CM.merge(ratings,on="tconst",how="left")
```

```python
modern_CMR.shape
```

```python
modern_CMR.isna().sum()
```

```python
modern_CMR = modern_CMR.dropna(axis=0,subset=['averageRating'],how='any')
modern_CMR.shape
```

```python
modern_CMR.tail()
```

```python
modern_CMR.to_csv('imdb_movies.csv')
```
