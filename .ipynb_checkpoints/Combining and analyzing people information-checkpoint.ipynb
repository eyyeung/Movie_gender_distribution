```python
import pandas as pd
```

# Data Visualization

```python
people = pd.read_csv('people.csv',index_col=['nconst'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(people.popularity)
```

```python
category_count = sns.countplot(x = 'category',
              data = people,
              order = people.category.value_counts().index)
category_count.set_xticklabels(category_count.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
