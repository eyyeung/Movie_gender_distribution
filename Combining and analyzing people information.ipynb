```python
movies = pd.read_csv('imdb_movies.csv',index_col='tconst')
```

```python
movies = movies.drop(columns=['directors','writers','originalTitle'])
```

```python
movies.head()
```

```python
test1 = pd.crosstab(index=[people.tconst,people.type], columns=people.gender)
```

```python
type(test1)
```

```python
test1.head()
```

```python
unstack_test = test1.unstack()
unstack_test = unstack_test.fillna(0)
```

```python
unstack_test[0].head()
```

```python
people_gender = pd.DataFrame(index=unstack_test.index, columns=['f_acting','f_creative','f_crew','m_acting','m_creative','m_crew'])
```

```python
people_gender.f_acting = unstack_test[1]['acting']
people_gender.f_creative = unstack_test[1]['creative']
people_gender.f_crew = unstack_test[1]['crew']

people_gender.m_acting = unstack_test[2]['acting']
people_gender.m_creative = unstack_test[2]['creative']
people_gender.m_crew = unstack_test[2]['crew']
```

```python
print(people_gender.isna().sum())
# no missing values - good!
```

```python
people_gender.head()
```

```python
movies_ppl = movies.merge(people_gender,on='tconst',how='inner')
```

```python
print(movies.shape)
print(people_gender.shape)
print(movies_ppl.shape)
```

```python
movies_ppl.head()
```

```python
movies_ppl['f_acting_frac'] = movies_ppl['f_acting']/(movies_ppl['f_acting']+movies_ppl['m_acting'])
movies_ppl['f_creative_frac'] = movies_ppl['f_creative']/(movies_ppl['f_creative']+movies_ppl['m_creative'])
movies_ppl['f_crew_frac'] = movies_ppl['f_crew']/(movies_ppl['f_crew']+movies_ppl['m_crew'])
```

```python
movies_ppl['total_ppl'] = movies_ppl['f_acting']+movies_ppl['f_creative']+movies_ppl['f_crew']+ movies_ppl['m_acting']+movies_ppl['m_creative']+movies_ppl['m_crew']
```

```python
movies_ppl['f_frac'] = (movies_ppl['f_acting']+movies_ppl['f_creative']+movies_ppl['f_crew'])/movies_ppl['total_ppl']
movies_ppl['m_frac'] = (movies_ppl['m_acting']+movies_ppl['m_creative']+movies_ppl['m_crew'])/movies_ppl['total_ppl']
```

```python
movies_ppl = movies_ppl.dropna(axis=0,subset=['runtimeMinutes','genres','f_frac'])
```

```python
movies_ppl = movies_ppl.fillna(0)
```

```python
movies_ppl.to_csv('movies_ppl_top10.csv')
```
